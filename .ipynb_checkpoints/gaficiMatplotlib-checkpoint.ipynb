```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 4 * np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# Usando matplotlib con apprccio OO (oject oriented - no pyplot)
fig, ax = plt.subplots(figsize=(6, 3)) # figsize = misura figura in pollici (serve per essere stampata) - 
#fig = figura - ax = oggetto Axes per creare un grafico (1 ax = 1 grafico)

ax.plot(x, y1, "b--", linewidth=5, label="sin(x)")
ax.plot(x, y2, "r-.", linewidth=5, label="cos(x)")
ax.set_title("un grafico a linea", fontsize=12)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
ax.grid()
plt.show()
```

grafico di x^2-3-4

```python
x = np.linspace(0, 1, 1000)
y1 = x**2
y2 = x**3
y3 = x**4
```

```python
fig, ax = plt.subplots(figsize=(3, 3)) 

ax.plot(x, y1, "b", label="$x^2$")
ax.plot(x, y2, "r", label="$x^3$")
ax.plot(x, y3, "g", label="$x^4$")
ax.set_title("un grafico a linea", fontsize=12)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
ax.grid()

plt.savefig("grafico.pdf", dpi=200)
plt.show()
```
