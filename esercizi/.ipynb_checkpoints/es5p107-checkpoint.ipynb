Considera un array monodimensionale di numeri e trova la posizione di tutti i massimi locali. Per esempio, dato l'array a = [0, 2, 1, 1, 3, 2], 2 e 3 sono massimi locali rispettivamente nelle posizioni 1 e 4.

```python
import numpy as np
```

```python
a = np.array([0, 2, 1, 1, 3, 2])
a
```

```python
aDx = np.roll(a, 1)
aSx = np.roll(a, -1)
```

```python
maxLocali = a[(a > aDx ) & (a > aSx)]
maxLocali
```

```python
posizioniMaxLocali = np.where((a > aDx ) & (a > aSx))
posizioniMaxLocali
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(a)
```
